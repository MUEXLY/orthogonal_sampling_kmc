```python
from analysis import DumpAnalyzer
import lzma
```

```python
APPROVAL_THRESHOLD = 0.98
```

```python
for i in range (1,5):
    with lzma.open(f'../analysis_many_dims/run_{i}/1000_1D_0.10_Sigma.txt.xz', 'tr') as f:
        analyzer = DumpAnalyzer(import_data_file=f)
        analyzer.standard_deviation = None
        r = analyzer.fit_line(step_size=100)
        
        print(f'run_{i} 1D, {r} {"Approved" if r >= APPROVAL_THRESHOLD else "Rejected"}')
        analyzer.generate_plot()
```
